```python
from tkinter import Tk, Label, PhotoImage, Frame, Text, GROOVE, Button, WORD
from tkinter import END
from tkinter.ttk import Combobox
from tkinter.filedialog import askdirectory
import pyttsx3
import speech_recognition as sr
import os
import threading  # Import threading module for concurrent execution

root = Tk()
root.title("VOICE")
root.geometry("900x450+100+100")
root.resizable(False, False)
root.configure(bg="#305065")

engine = pyttsx3.init()
r = sr.Recognizer()
stop_listening = False  # Variable to control the continuous listening


def speaknow():
    text = text_area.get(1.0, END)
    gender = gender_combobox.get()
    speed = speed_combobox.get()
    voices = engine.getProperty("voices")

    def setvoice():
        if gender == "MALE":
            engine.setProperty("voice", voices[0].id)
        else:
            engine.setProperty('voice', voices[1].id)

    if text:
        setvoice()
        if speed == "FAST":
            engine.setProperty("rate", 250)
        elif speed == "NORMAL":
            engine.setProperty("rate", 150)
        else:
            engine.setProperty("rate", 60)

        engine.say(text)
        engine.runAndWait()

def download():
    text = text_area.get(1.0, END)
    gender = gender_combobox.get()
    speed = speed_combobox.get()
    voices = engine.getProperty("voices")

    def setvoice():
        if gender == "MALE":
            engine.setProperty("voice", voices[0].id)
        else:
            engine.setProperty('voice', voices[1].id)

    if text:
        setvoice()
        if speed == "FAST":
            engine.setProperty("rate", 250)
        elif speed == "NORMAL":
            engine.setProperty("rate", 150)
        else:
            engine.setProperty("rate", 60)

        path = askdirectory()
        os.chdir(path)
        engine.save_to_file(text, "text.mp3")
        engine.runAndWait()

def continuous_listening():
    global stop_listening
    while not stop_listening:
        try:
            with sr.Microphone() as source:
                r.adjust_for_ambient_noise(source, duration=0.2)
                print("Say something:")
                audio = r.listen(source)
                text = r.recognize_google(audio, language="en-US")
                print("Speech to Text:", text)
                text_area.insert(END, "\n" + text)
                root.update()  # Update the GUI to show the recognized text immediately
        except sr.RequestError as e:
            print(f"Could not request results: {e}")
        except sr.UnknownValueError:
            print("Unknown error occurred")


def start_listening():
    global stop_listening
    stop_button["state"] = "normal"  # Enable the "Stop Listening" button
    start_button["state"] = "disabled"  # Disable the "Start Listening" button
    stop_listening = False
    threading.Thread(target=continuous_listening).start()  # Start continuous listening in a separate thread

def stop_listening_func():
    global stop_listening
    stop_listening = True
    stop_button["state"] = "disabled"  # Disable the "Stop Listening" button
    start_button["state"] = "normal"  # Enable the "Start Listening" button

image_icon = PhotoImage(file="icone-de-microphone-rouge.png")
root.iconphoto(False, image_icon)

Top_frame = Frame(root, bg="white", width=900, height=100)
Top_frame.place(x=0, y=0)

original = PhotoImage(file="speaking-talking-png-14.png")
Logo = original.subsample(3, 3)
Label(Top_frame, image=Logo, bg="white").place(x=10, y=5)

Label(Top_frame, text="TEXT <---> SPEECH", font="arial 20 bold", bg="white", fg="black").place(x=100, y=30)

text_area = Text(root, font="Roboto 20", bg="white", relief=GROOVE, wrap=WORD)
text_area.place(x=10, y=150, width=500, height=250)

Label(root, text="VOICE", font="arial 15 bold", bg="#305065", fg="white").place(x=580, y=160)
Label(root, text="SPEED", font="arial 15 bold", bg="#305065", fg="white").place(x=760, y=160)

gender_combobox = Combobox(root, values=["MALE", "FEMALE"], font="arial 14", state="r", width=10)
gender_combobox.place(x=550, y=200)
gender_combobox.set("MALE")

speed_combobox = Combobox(root, values=["FAST", "NORMAL", "SLOW"], font="arial 14", state="r", width=10)
speed_combobox.place(x=730, y=200)
speed_combobox.set("NORMAL")

zero = PhotoImage(file="icone-de-microphone-rouge.png")
imageicon = zero.subsample(10, 10)
btn = Button(root, text="Speak", compound="none", image=imageicon, width=130, font="arial 14 bold", command=speaknow)
btn.place(x=550, y=280)

one = PhotoImage(file="save.png")
imageicon2 = one.subsample(10, 10)
save = Button(root, text="Save", compound="none", image=imageicon2, width=130, bg="#39c790", font="arial 14 bold", command=download)
save.place(x=730, y=280)

start_button = Button(root, text="Speak", width=15, font="arial 12 bold", command=start_listening)
start_button.place(x=550, y=360)

stop_button = Button(root, text="Stop-speaking", width=15, font="arial 12 bold", command=stop_listening_func, state="disabled")
stop_button.place(x=730, y=360)

root.mainloop()
```
